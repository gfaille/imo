```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import missingno as msno
import sklearn
from sklearn.model_selection import train_test_split
from matplotlib.colors import ListedColormap
from sklearn.linear_model import Lasso, ElasticNet
from sklearn.preprocessing import MinMaxScaler, RobustScaler, PolynomialFeatures, StandardScaler,  Binarizer
from sklearn.pipeline import make_pipeline
import pickle

pd.options.display.max_rows = 150
pd.options.display.max_columns = 90
```

```python
df = pd.read_csv("immobilier2.csv")
df
```

```python
X = df.drop(["price","sqft_living15", "sqft_lot15", "lat", "long"], axis=1)
Y = df["price"]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=0)

reg = make_pipeline(MinMaxScaler(), PolynomialFeatures(degree=2), Lasso(alpha=13.59))
reg.fit(X_train, Y_train)
# X = features / y = target
```

```python
reg.score(X_test, Y_test)
```
